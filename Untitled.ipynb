# Spliting data to test data and training Data

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split(x,y,test_size=0.3)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

# The datasets has been split into training data(training data input sets and training data output set) and testing data(testing data input sets and testing data output set) 

```python
# Now we will Scale the Values
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)#first it will understand the data and then scale it and will return an array for that!!...
```

```python
x_train
```

```python
x_test=scaler.transform(x_test)
```

```python
x_test
```

```python
# Now we will use the LOgistic ALgo to train our model!!..
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf=LogisticRegression()
```

```python
clf.fit(x_train,y_train.values.ravel())
```

```python
y_predict=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_predict)
```

```python
 import pickle
```

```python
pickle.dump(clf,open("model1.pkl",'wb'))
```
